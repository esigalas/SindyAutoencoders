```python
# load test and train datasets
# training_model_data = np.load("flat_model_train_data_target.npy") #get_lorenz_data(1024, noise_strength=noise_strength)
# validation_model_data = np.load("flat_model_test_data_target.npy") #get_lorenz_data(20, noise_strength=noise_strength)
# training_model_data_derivatives = np.load("flat_model_train_data_derivatives_target.npy")
# validation_model_data_derivatives = np.load("flat_model_test_data_derivatives_target.npy")
training_model_data = np.load("data/new_flat_model_train_data_target.npy")
validation_model_data = np.load("data/new_flat_model_test_data_target.npy")
training_model_data_derivatives = np.load("data/new_flat_model_train_data_derivatives_target.npy")
validation_model_data_derivatives = np.load("data/new_flat_model_test_data_derivatives_target.npy")
# tt = get_lorenz_data(1024, noise_strength=noise_strength)
training_data ={}
validation_data = {}
training_data['x'] = training_model_data
training_data['dx'] = training_model_data_derivatives
validation_data['x'] = validation_model_data
validation_data['dx'] = validation_model_data_derivatives
training_model_data.shape
```

# Set up model and training parameters

```python
params = {}

params['input_dim'] = 120
params['latent_dim'] = 2
params['model_order'] = 1
params['poly_order'] = 5
params['include_sine'] = True
params['library_dim'] = library_size(params['latent_dim'], params['poly_order'], params['include_sine'], True)

# sequential thresholding parameters
params['sequential_thresholding'] = True
params['coefficient_threshold'] = 0.6
params['threshold_frequency'] = 500
params['coefficient_mask'] = np.ones((params['library_dim'], params['latent_dim']))
params['coefficient_initialization'] = 'constant'

# loss function weighting
params['loss_weight_decoder'] = 1
params['loss_weight_sindy_z'] = 0.9
params['loss_weight_sindy_x'] = 0.9
params['loss_weight_sindy_regularization'] = 1e-4

params['activation'] = 'relu'
params['widths'] = [64,32]

# training parameters
params['epoch_size'] = training_data['x'].shape[0]
params['batch_size'] = 1024
params['learning_rate'] = 1e-3

params['data_path'] = os.getcwd() + '/'
params['print_progress'] = True
params['print_frequency'] = 100

# training time cutoffs
params['max_epochs'] = 5001
params['refinement_epochs'] = 0
```

# Run training experiments

```python
num_experiments = 1
df = pd.DataFrame()
np.random.seed(1)


for i in range(num_experiments):
    print('EXPERIMENT %d' % i)

    params['coefficient_mask'] = np.ones((params['library_dim'], params['latent_dim']))

    params['save_name'] = 'models/bump_attractor_order'+str(params['poly_order'])+'_'+str(params['activation'])+'_epochs'+str(params['max_epochs'])+ datetime.datetime.now().strftime("%Y_%m_%d_%H_%M_%S_%f")

    tf.reset_default_graph()
    tf.compat.v1.set_random_seed(1)
    results_dict = train_network(training_data, validation_data, params)
    df = df.append({**results_dict, **params}, ignore_index=True)

df.to_pickle('models/experiment_results_' + datetime.datetime.now().strftime("%Y%m%d%H%M") + '.pkl')
```
